```python
#import neccessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
#load up required csv file

pd.set_option('display.max_columns', None) 

df = pd.read_csv('healthmopupandbaselinenmisfacility.csv')
df.head(30)
```

```python
df.info()
```

```python
df.describe(percentiles=[0.85,0.90,0.95])
```

```python
columns=['num_chews_fulltime','num_nurses_fulltime','num_nursemidwives_fulltime','num_doctors_fulltime']
percentile=df[columns].quantile(0.99)
print(percentile)
```

```python
# Outliers detection and removal

for i in columns:
    upper_limit = percentile[i]  
    count = (df[i] >= upper_limit).sum()  
    print(f"{i}: {count}")
    
for i in columns:
    upper_limit = percentile[i] 
    df = df[df[i] < upper_limit]
    
df.shape
```

```python
# Sorting and removing null values

null_values=df.isnull().sum()
non_null_values=df.notnull().sum()

ratio=null_values/non_null_values
print(ratio)
```

```python
# Remove Unneeded variables

variables_to_remove = [
    'facility_name', 'date_of_survey', 'ward', 'sector', 
    'formhub_photo_id', 'gps', 'survey_id', 'unique_lga', 
    'latitude', 'longitude'
]

df = df.drop(columns=variables_to_remove)
```

```python
# Replace null values with mode for specific variables

variables_replace_mode = [
    'phcn_electricity', 'child_health_measles_immun_calc', 
    'management', 'improved_water_supply', 'vaccines_fridge_freezer'
]

for col in variables_replace_mode:
    mode_value = df[col].mode()[0]
    df[col] = df[col].fillna(mode_value)
```

```python
# Replace null values with mean for specified variables

variables_replace_mean = [
    'num_chews_fulltime', 'num_nurses_fulltime', 
    'num_nursemidwives_fulltime', 'num_doctors_fulltime'
]

for col in variables_replace_mean:
    mean_value = df[col].mean()
    df[col] = df[col].fillna(mean_value)
```

```python
# Remove rows with null values from specified variables

variables_remove_nulls = [
    'facility_type_display', 'maternal_health_delivery_services', 
    'emergency_transport', 'skilled_birth_attendant', 'c_section_yn', 
    'facility_id', 'community', 'management', 'improved_sanitation', 
    'antenatal_care_yn', 'family_planning_yn', 'malaria_treatment_artemisinin'
]

df = df.dropna(subset=variables_remove_nulls)
```

```python
# Encode nominal variables

variables_to_encode = [
    'facility_type_display', 'maternal_health_delivery_services', 
    'emergency_transport', 'skilled_birth_attendant', 'phcn_electricity', 
    'c_section_yn', 'child_health_measles_immun_calc', 'facility_id', 
    'community', 'management', 'improved_water_supply', 
    'improved_sanitation', 'vaccines_fridge_freezer', 'antenatal_care_yn', 
    'family_planning_yn', 'malaria_treatment_artemisinin'
]

label_encoders = {}
for col in variables_to_encode:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le  # Save encoders if you need to inverse-transform later

# Display the first few rows of the cleaned DataFrame
df.head()
```
